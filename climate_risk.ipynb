```python
#import necessary libraries
from pypdf import PdfReader
```

```python
#define directories
reading_directory = '/Users/pelumioluwaabiola/Downloads/Climate Risk'
```

```python
#read pdf file
report_climate_related = []
document_name = '/190618-climate-related-information-reporting-guidelines_en.pdf'
reader = PdfReader(reading_directory + document_name)
number_of_pages = len(reader.pages)
page = reader.pages[2]
text = page.extract_text()

for i in range(0, number_of_pages):
    page = reader.pages[i]
    text = page.extract_text()
    report_climate_related.append(text)

print(report_climate_related)
```

```python
def clean_pdf(text):
    # If the first character is a digit, remove it
    for _ in range(4):
        if text and text[0].isdigit():
            text = text[1:]
    #remove '\n' (newlines)
    text = text.replace('\n', '  ')
    # Remove '\x0c' (form feed/new page)
    text = text.replace('\x0c', ' ')
    # Remove '\xa0' (non-breaking space)
    text = text.replace('\xa0', ' ')
    # Remove '\xa0' (non-breaking space)
    text = text.replace('\uf0b7 ', ' ')
    #remove all unnecessary spaces
    text = ' '.join(text.split())

    return text
```

```python
#clean pdf
clean_report_climate_related = []
for i in range(len(report_climate_related)):
    text = clean_pdf(report_climate_related[i])
    clean_report_climate_related.append(text)

clean_report_climate_related
```
